## Q1. 寫一個函式，抓取 Google Finance 的股價歷史資料

```python
import pandas as pd
import io
from datetime import datetime
import requests
```

```python
def goog_data_reader(stock, start_date):
    url = "http://finance.google.com/finance/historical"

    params = {}
    params['q'] = stock
    params['startdate'] = start_date
    params['output'] = 'csv'

    r = requests.get(url, params=params)
    rawData = pd.read_csv(io.StringIO(r.text)) #轉成dataframe

    return rawData
```

```python
goog_data_reader('TSLA', '2015-01-01')
```
